```python
def actualizar_modelo_sugerido(df):
    """
    Función que actualiza la columna modelo_sugerido usando el valor de Aux
    cuando modelo_sugerido es igual a 'No encontrado' y agrega una columna
    que indica si el registro fue modificado. También crea una columna modelo_final
    con los caracteres especiales reemplazados por espacios
    """
    # Crear una copia del dataframe para no modificar el original
    df_actualizado = df.copy()

    # Crear columna para indicar si fue modificado
    df_actualizado['fue_modificado'] = df_actualizado['modelo_sugerido'] == 'No encontrado'

    # Aplicar la lógica: si modelo_sugerido es 'No encontrado', usar el valor de Aux
    df_actualizado['modelo_sugerido'] = df_actualizado.apply(
        lambda row: row['Aux'] if row['modelo_sugerido'] == 'No encontrado' else row['modelo_sugerido'],
        axis=1
    )

    # Crear columna modelo_final reemplazando caracteres especiales por espacios
    df_actualizado['modelo_final'] = df_actualizado['modelo_sugerido'].str.replace('-', ' ', regex=False)

    return df_actualizado
```

```python
# Aplicar la función a los dataframes
df_mx_actualizado = actualizar_modelo_sugerido(df_mx)
df_co_actualizado = actualizar_modelo_sugerido(df_co)
df_cl_actualizado = actualizar_modelo_sugerido(df_cl)
```

```python
# Recopilar TODOS los modelos originales de todos los países
todos_modelos_originales = []

# Colombia
if not df_co_actualizado.empty:
    co_modelos = df_co_actualizado[['Modelo', 'modelo_final', 'code_modelos_agrupados']].copy()
    co_modelos.columns = ['modelo_original', 'modelo_final', 'code_modelos_agrupados']
    co_modelos['pais'] = 'CO'
    todos_modelos_originales.append(co_modelos)

# México
if not df_mx_actualizado.empty:
    mx_modelos = df_mx_actualizado[['Modelo', 'modelo_final', 'code_modelos_agrupados']].copy()
    mx_modelos.columns = ['modelo_original', 'modelo_final', 'code_modelos_agrupados']
    mx_modelos['pais'] = 'MX'
    todos_modelos_originales.append(mx_modelos)

# Chile
if not df_cl_actualizado.empty:
    cl_modelos = df_cl_actualizado[['Modelo', 'modelo_final', 'code_modelos_agrupados']].copy()
    cl_modelos.columns = ['modelo_original', 'modelo_final', 'code_modelos_agrupados']
    cl_modelos['pais'] = 'CL'
    todos_modelos_originales.append(cl_modelos)

# Concatenar todos
df_todos_modelos = pd.concat(todos_modelos_originales, ignore_index=True)
```

```python
# Crear una clave única que combine modelo_original Y modelo_final
def crear_clave_unica(row):
    """Crea una clave única para cada modelo"""
    original = str(row['modelo_original']).strip()
    final = str(row['modelo_final']).strip()

    # Si modelo_final es válido, usarlo; sino usar original
    if final and final != 'nan' and final != 'No encontrado':
        return final
    else:
        return original

df_todos_modelos['clave_modelo'] = df_todos_modelos.apply(crear_clave_unica, axis=1)

# Obtener todas las claves únicas
claves_unicas = df_todos_modelos['clave_modelo'].unique()
print(f"Total de modelos únicos: {len(claves_unicas)}")

# Crear matriz pivoteada
df_unificado = df_todos_modelos.pivot_table(
    index='clave_modelo',
    columns='pais',
    values=['modelo_original', 'code_modelos_agrupados'],
    aggfunc='first'  # En caso de duplicados, tomar el primero
).reset_index()

# Aplanar columnas multi-nivel
df_unificado.columns = [
    'modelo_unificado',
    'code_cl', 'code_co', 'code_mx',  # codes
    'original_cl', 'original_co', 'original_mx'  # originales
]
```

```python
df_unificado_excel = df_unificado.fillna("No disponible")
```

```python
df_unificado_excel['disponible_en'] = df_unificado_excel.apply(
    lambda row: ', '.join([
        'Chile' if pd.notna(row['code_cl']) and row['code_cl'] != 'No disponible' else '',
        'México' if pd.notna(row['code_mx']) and row['code_mx'] != 'No disponible' else '',
        'Colombia' if pd.notna(row['code_co']) and row['code_co'] != 'No disponible' else ''
    ]).strip(', '), axis=1
)

# Función para verificar si necesita homologación y en qué países
def verificar_homologacion(row):
    modelo_final = row['modelo_final']
    modelo_cl = row['Modelo_cl']
    modelo_mx = row['Modelo_mx']
    modelo_co = row['Modelo_co']

    paises_para_homologar = []

    # Verificar Chile
    if modelo_cl != 'No disponible' and modelo_cl != modelo_final:
        paises_para_homologar.append('Chile')

    # Verificar México
    if modelo_mx != 'No disponible' and modelo_mx != modelo_final:
        paises_para_homologar.append('México')

    # Verificar Colombia
    if modelo_co != 'No disponible' and modelo_co != modelo_final:
        paises_para_homologar.append('Colombia')

    if paises_para_homologar:
        return ', '.join(paises_para_homologar)
    else:
        return 'No necesita'

# Aplicar la función para determinar homologación
df_unificado_excel['necesita_homologacion'] = df_unificado_excel.apply(verificar_homologacion, axis=1)
```
